```python
import pandas as pd
import requests
import json

# URLs for your datasets
train_url = 'https://raw.githubusercontent.com/UpatelPenn/DS340W-Project/main/RE-Data/BC5CDR/train.json'
test_url = 'https://raw.githubusercontent.com/UpatelPenn/DS340W-Project/main/RE-Data/BC5CDR/test.json'
valid_url = 'https://raw.githubusercontent.com/UpatelPenn/DS340W-Project/main/RE-Data/BC5CDR/valid.json'

def load_and_clean_data(url):
    try:
        # Fetch the data
        response = requests.get(url)
        response.raise_for_status()  # Check for HTTP errors

        # Load JSON data
        data = json.loads(response.text)

        # Prepare lists to hold the extracted information
        relations = []

        for entry in data.values():
            for relation in entry.get('relations', []):
                chemical = relation['chemical']
                disease = relation['disease']
                relations.append((chemical, disease))

        # Convert the relations list to a DataFrame
        df = pd.DataFrame(relations, columns=['chemical', 'disease'])

        # Limit to the first 40000 rows
        df = df.head(40000)

        return df

    except Exception as e:
        print(f"Error processing data from {url}: {e}")
        return None

# Load and clean the data from each dataset
train_data = load_and_clean_data(train_url)
test_data = load_and_clean_data(test_url)
valid_data = load_and_clean_data(valid_url)

# Output the first few rows of each cleaned dataset for inspection
print("Train Data:")
print(train_data.head())
print("\nTest Data:")
print(test_data.head())
print("\nValidation Data:")
print(valid_data.head())

# Function to calculate metrics
def calculate_metrics(true_labels, predicted_labels):
    precision = precision_score(true_labels, predicted_labels, average='weighted', zero_division=0)
    recall = recall_score(true_labels, predicted_labels, average='weighted', zero_division=0)
    f1 = f1_score(true_labels, predicted_labels, average='weighted', zero_division=0)

    return precision, recall, f1

# Function to create simulated predictions
def simulate_predictions(true_labels):
    import numpy as np
    unique_labels = np.unique(true_labels)
    # Randomly assign labels for demonstration purposes
    return np.random.choice(unique_labels, size=len(true_labels))

# Calculate metrics for each dataset
for name, data in zip(['Train', 'Test', 'Validation'], [train_data, test_data, valid_data]):
    # Create a true label for simplicity; in real scenarios, you would have actual predictions
    true_labels = data['disease'].values  # Assuming the disease column is the ground truth

    # Simulate predicted labels
    predicted_labels = simulate_predictions(true_labels)

    # Calculate metrics
    precision, recall, f1 = calculate_metrics(true_labels, predicted_labels)

    # Print results
    print(f"\n{name} Data Metrics:")
    print(f'Precision: {precision:.4f}')
    print(f'Recall: {recall:.4f}')
    print(f'F1 Score: {f1:.4f}')
```
